```python
def readFile(filename, pthreshold):
    file = open(filename)
    firstLine = file.readline().lower()
    splitUp = firstLine.split()
    pval_index = 0
    id_index = 0
    count = 0
    for i in splitUp:
        if (i == 'rsid' or i == 'snp' or i == 'marker' or i == 'rsids'):
            id_index = count
        if (i == 'p' or (i.find('p')!=-1 and i.find('value')!=-1)):
            pval_index = count
        count+=1
        drData = {} # {'sampleRsid': 0}
        
    with open(filename) as fp:
        lineNum = 0
        for line in fp:
            if(lineNum != 0):
                splitLine = line.split()
                id = '';
                drData[splitLine[id_index]] = float(splitLine[pval_index])
            lineNum+=1
            
    return {k:v for (k,v) in drData.items() if float(v) < pthreshold}
```

```python
def returnIds(dict1, dict2):
    
    keys1 = dict1.keys()
    keys2 = dict2.keys()
    
    return set(keys1) - (set(keys1) - set(keys2))
```
